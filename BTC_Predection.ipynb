```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/AchrefHamdi/-BTC-USD--Predection/main/BTC-USD.csv")
```

```python
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
data.corr()
sns.heatmap(data.corr())
plt.show()
```

```python
figure = go.Figure(data=[go.Candlestick(x=data["Date"],
                                        open=data["Open"], high=data["High"],
                                        low=data["Low"], close=data["Close"])])
figure.update_layout(title = "BTC Price", xaxis_rangeslider_visible=False)
figure.show()
```

```python
from autots import AutoTS
model = AutoTS(forecast_length=10, frequency='infer', ensemble='simple')
model = model.fit(data, date_col='Date', value_col='Close', id_col=None)
prediction = model.predict()
forecast = prediction.forecast
print(forecast)
```
